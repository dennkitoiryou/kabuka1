```
#MB5 = Make_base_dataset_min(sales_df=sales_df, days=1, id_no=1)の場合
import pandas as pd
import itertools
import numpy as np

calendar_df = pd.read_csv("/content/drive/My Drive/calendar_min.csv")
sales_df = pd.read_csv("/content/drive/My Drive/practice/practice_train.csv")

#print("calendar_df")
#print(calendar_df)
print("sales_df")
print(sales_df)


class Make_base_dataset_min():
    def __init__(self, sales_df, days, id_no):
        self.sales_df = sales_df
        self.id_no = id_no
        self.days = days

    def copy_feature(self, feature_array):
        feature_array = feature_array[:self.id_no].tolist()
        print("feature_array_[:self.id_no]_.tolist()を実行した場合")
        print(feature_array)
        print("---------------------------------------------")

        #print("for_i_in_range_int_self.days_を実行した場合")
        #print(for i in range(int(self.days))

        print("---------------------------------------------")

        print("feature_array for_i_in_range(int(self.days))を実行した場合")
        print([feature_array for i in range(int(self.days))])

        print("---------------------------------------------")
        print("Movement when returning　return listを実行した場合")
        print(itertools.chain.from_iterable([feature_array for i in range(int(self.days))]))
        return list(itertools.chain.from_iterable([feature_array for i in range(int(self.days))]))

    def make_id_feature(self):
        print(self.copy_feature(self.sales_df["id"].values))
        return self.copy_feature(self.sales_df["id"].values)
    
    def make_id_feature_p(self):
        MB_5_p=self.sales_df["id"].values
        print(MB_5_p)

    def make_id_feature_p2(self):
        MB_5_p2=self.sales_df["item_id"].values
        print(MB_5_p2)

MB5 = Make_base_dataset_min(sales_df=sales_df, days=1, id_no=1)
print("MB5.make_id_feature(),idの列をスライスする")
MB5.make_id_feature()
print("---------------------------------------------")
print("MB5.make_id_feature_p(),self.sales_df,idのvalues:値を抜いて表示させる")
MB5.make_id_feature_p()

class Export_pickle():

    def __init__(self):
        """特徴量の命名方法　：　作った人のイニシャル_元の特徴量_内容（例 : F_month_categorized_quarter 月の特徴量を四半期でカテゴライズ）"""
        #作り方がわからないです↓　files name : str : ex:F_month_categorized_quarter.pickle
    def export_pickle_files (self,feature_list, files_name):
        # パス変えてね
        feature_df = pd.DataFrame(feature_list)
        feature_df.to_pickle('/content/drive/My Drive/practice/{}'.format(files_name))

#def main():
    #MB = Make_base_dataset(sales_df=sales_df, days=1, id_no=1)
    #Ex = Export_pickle()
    #Ex.export_pickle_files(MB.make_id_feature(), "F_id_base_min_endo.pickle")
    
#if __name__ == "__main__":
#    main()
```
